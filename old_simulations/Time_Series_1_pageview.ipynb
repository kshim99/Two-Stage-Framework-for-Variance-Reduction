```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import statsmodels.api as sm
import matplotlib.pyplot as plt
from datetime import date
from datetime import datetime
import random
import itertools
import io
```

```python
data = pd.read_csv("../AB/page_visits_assignment.csv")
data.head(30)
```

```python
# reformat the date and turn into datetime obj
data["Date"] = pd.to_datetime(data["Date"], format='mixed')
```

```python
# extract only relevant columns: time and outcome variable
df = data[["Date","Visits"]]
df = df.rename(columns={'Date': 'x', 'Visits': 'y'})
# remove data past 2019 november as it is unusually volatile
df = df[df['x'] < '2019-11-01']
# Separate train, validation, test data
n = df.shape[0]
train_prop = 0.95
val_prop = 0.025
test_prop = 1 - train_prop - val_prop
train_df = df[:int(n*train_prop)]
val_df = df[int(n*train_prop):int(n*(1-test_prop))]
test_df = df[int(n*(1-test_prop)):]

# plot
plt.rcParams["figure.figsize"] = (10,10)
x_range = [min(df['x']), max(df['x'])]
y_range = [min(df['y']), max(df['y'])]
plt.plot(train_df['x'], train_df['y'], color='black',linewidth=0.5,label='train_data')
plt.plot(val_df['x'], val_df['y'], color='blue',linewidth=0.5,label='val_data')
plt.plot(test_df['x'], test_df['y'], color='red',linewidth=0.5,label='test_data')
ax = plt.gca()
ax.set_xlim(x_range)
ax.set_ylim(y_range)
ax.set_title("Time Series of Page Visits")
ax.set_xlabel("Time")
ax.set_ylabel("Page Visits")
plt.show()
```

```python
# variance seems to be increasing over time. Use log transform for constant variance.
df['y'] = np.log(df['y'])
train_df = df[:int(n*train_prop)]
val_df = df[int(n*train_prop):int(n*(1-test_prop))]
test_df = df[int(n*(1-test_prop)):]

# plot
plt.rcParams["figure.figsize"] = (10,10)
x_range = [min(df['x']), max(df['x'])]
y_range = [min(df['y']), max(df['y'])]
plt.plot(train_df['x'], train_df['y'], color='black',linewidth=0.5,label='train_data')
plt.plot(val_df['x'], val_df['y'], color='blue',linewidth=0.5,label='val_data')
plt.plot(test_df['x'], test_df['y'], color='red',linewidth=0.5,label='test_data')
ax = plt.gca()
ax.set_xlim(x_range)
ax.set_ylim(y_range)
ax.set_title("Time Series of Log Transformed Page Visits")
ax.set_xlabel("Time")
ax.set_ylabel("Log Transformed Page Visits")
plt.show()
```

```python
# Follow box jenkins methodology to fit a sarima model
# ONLY use TRAIN data
fig, axes = plt.subplots(1, 2, figsize=(15,4))
fig = sm.graphics.tsa.plot_acf(train_df['y'], ax=axes[0])
fig = sm.graphics.tsa.plot_pacf(train_df['y'], ax=axes[1])
```

```python
# Choose parameters to try fitting the model with
p = [0,1]
q = [1,6]
d = [1]
P = [0,1]
Q = [0,1]
D = [0,1]
S = [7]
param1 = list(itertools.product(p,d,q))
param2 = list(itertools.product(P,D,Q,S))
best_param1 = param1[0]
best_param2 = param2[0]
for i in range(len(param1)):
    for j in range(len(param2)):
        #print(i,j)
        model = sm.tsa.statespace.SARIMAX(train_df['y'], order = param1[i], seasonal_order = param2[j])
        result = model.fit()
        tb = result.summary().tables[0].as_html()
        readable_tb = pd.read_html(io.StringIO(tb),header=None,index_col=None)[0]
        cur_aic = readable_tb[3][2]
        if i == 0 and j == 0:
            best_aic = cur_aic
        elif cur_aic < best_aic:
            best_aic = cur_aic
            best_param1 = param1[i]
            best_param2 = param2[j]
best_aic, best_param1, best_param2
```

```python
# fit the best model and check residuals
opt_model = sm.tsa.statespace.SARIMAX(train_df['y'], order = best_param1, seasonal_order = best_param2)
opt_res = opt_model.fit()
# sarimax uses kalman filter which uses the first period to find periodicity, hence fit is poor in the first 7 obs
# consider the fit and residuals after the first 7
fig, ax = plt.subplots(2,1)
ax[0].plot(train_df['x'][7:], train_df['y'][7:], color='black',linewidth=0.5,label='train_data')
ax[0].plot(train_df['x'][7:], opt_res.fittedvalues[7:], color='red',linewidth=0.5, label='fitted_data')
ax[0].set_title("Log Transformed Page Visits of Train Data and Fitted Values")
ax[1].hist(opt_res.resid[7:])
ax[1].set_title("Fitted Residuals Distribution")
plt.show()
```
