```python
page_visits['Date']
# there are two different date formatting - need customized approach to standardize it
```

```python
def autoconvert_datetime(value):
    formats = ['%m-%d-%Y','%m/%d/%Y']  # formats to try
    result_format = '%d-%m-%Y'  # output format
    for dt_format in formats:
        try:
            dt_obj = dt.strptime(value, dt_format)
            return dt.strptime(dt_obj.strftime(result_format), result_format)
        except Exception as e:  # throws exception when format doesn't match
            pass
    return value  # let it be if it doesn't match

page_visits['Date'] = page_visits['Date'].apply(autoconvert_datetime)
page_visits['Date']
```

```python
# make a copy of clean data with dates as row index
data = page_visits[['Date','Visits']]
data = data.set_index('Date')
data = data['Visits'].resample('MS').mean()
# lets plot the data
data.plot(figsize=(15,6))
plt.show()
```

```python
# decomposition of time series
decomposition = sm.tsa.seasonal_decompose(data, model = 'additive')
fig = decomposition.plot()
plt.show()
# can see correlation in residuals, need to remove it with time series models 
```
